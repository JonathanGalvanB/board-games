```python
import prince  # MCA library

# Sample DataFrame with boolean categorical variables
data = {
    'Has_Pet': [1, 0, 1, 1, 0],
    'Likes_Sports': [0, 1, 0, 1, 1],
    'Owns_Car': [1, 1, 0, 0, 1]
}

df = pd.DataFrame(data)

# Apply MCA
mca = prince.MCA(n_components = 2, random_state=42)
mca.fit(df)  # Fit the model

# Transform data into lower-dimensional space
df_mca = mca.transform(df)

# Display the transformed coordinates
print(df_mca)
```
