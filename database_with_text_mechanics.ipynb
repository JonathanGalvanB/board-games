```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("database_1000.json")
```

```python
df['New_Description'] = df.apply(
    lambda row: f"{row['Description']} \n The mechanics are: {', '.join(row['mechanics'])}.",
    axis=1
)
```

```python
df.to_json('database_1000_full_mechanics.json', orient='records', indent=4)
```
